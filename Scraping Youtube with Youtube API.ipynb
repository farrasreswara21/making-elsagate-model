# Search with Function

```python
def search(q):
    #searching parameters
    req = youtube.search().list(
    part = 'snippet',
    q = q, 
    maxResults = 50)
    
    #start searching (-100 quotas) 
    response = req.execute()
    
    #show nextPageToken
    print('use this token to go to the next page:', response['nextPageToken'])
    
    #processing raw data
    judul = []
    gambar = []
    for i in range(len(response['items'])):
        judul.append(response['items'][i]['snippet']['title'])
        gambar.append(response['items'][i]['snippet']['thumbnails']['default']['url'])
    
    clean_data = pd.DataFrame({'x1':judul,
                                'x2':gambar})
    return clean_data 
```

```python
def search_next_page(q, next_page_token):
    #searching parameters
    req = youtube.search().list(
    part = 'snippet',
    q = q, 
    maxResults = 50,
    pageToken = next_page_token
    )
    
    #start searching (-100 quotas) 
    response = req.execute()
    
    #show nextPageToken
    print('use this token to go to the next page:', response['nextPageToken'])
    
    #processing raw data
    judul = []
    gambar = []
    for i in range(len(response['items'])):
        judul.append(response['items'][i]['snippet']['title'])
        gambar.append(response['items'][i]['snippet']['thumbnails']['default']['url'])
    
    clean_data = pd.DataFrame({'x1':judul,
                                'x2':gambar})
    return clean_data 
```

# Make Dataset

```python
def get_youtube_data(num_of_data, kind, query=None, channel=None):
    titles = []
    thumbnails = []
    next_token = ''
    
    for i in tqdm(range(1, num_of_data, 50)):
        if len(titles) != num_of_data:
            get_data = youtube.search().list(part = 'snippet', q = query, maxResults = 50, pageToken = next_token, channelId = channel).execute()
            next_token = get_data['nextPageToken']
        
            for i in range(len(get_data['items'])):
                title = get_data['items'][i]['snippet']['title']
                thumbnail = get_data['items'][i]['snippet']['thumbnails']['default']['url']
                thumbnail_path = urllib.request.urlretrieve(thumbnail, f"{kind}/{thumbnail.split('/')[4]}.png")
                thumbnail_path = str(thumbnail_path).split("'")[1]
                
                titles.append(title)
                thumbnails.append(thumbnail_path)
                
        elif len(titles) == num_of_data:
            break            
            
    df = pd.DataFrame({'x1':titles, 'x2':thumbnails})
    if kind == 'elsagate':
        df['y'] = 1
    elif kind == 'not_elsagate':
        df['y'] = 0
        
    return df.to_csv(f'{kind}.csv', index=False)
    
```

```python
def get_youtube_data_from_many_source(num_of_data, kind, query=None, channel=None):
    titles = []
    thumbnails = []
    next_token = ''
    
    for q in tqdm(query):
        for i in tqdm(range(1, num_of_data, 50)):
            if len(titles) != num_of_data:
                get_data = youtube.search().list(part = 'snippet', q = q, maxResults = 50, pageToken = next_token, channelId = channel).execute()
                next_token = get_data['nextPageToken']
            
                for i in range(len(get_data['items'])):
                    title = get_data['items'][i]['snippet']['title']
                    thumbnail = get_data['items'][i]['snippet']['thumbnails']['default']['url']
                    thumbnail_path = urllib.request.urlretrieve(thumbnail, f"{kind}/{thumbnail.split('/')[4]}.png")
                    thumbnail_path = str(thumbnail_path).split("'")[1]
                    
                    titles.append(title)
                    thumbnails.append(thumbnail_path)
                    
            elif len(titles) == num_of_data:
                break            
            
    df = pd.DataFrame({'x1':titles, 'x2':thumbnails})
    if kind == 'elsagate':
        df['y'] = 1
    elif kind == 'not_elsagate':
        df['y'] = 0
        
    return df.to_csv(f'{kind}_data.csv', index=False)
```
